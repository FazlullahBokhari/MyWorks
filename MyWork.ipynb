```python
import sqlite3 as s
class BagOfWords:
    
    def readFile(self,file):
        '''
        return words from file
        This function use for reading the respective file
        we have to pass the path 
            
        '''
        try:
            data = open(file,'r')
            read_words = data.read()
            sep_word = read_words.split(sep='\n')
        except Exception as e:
            return e
        return sep_word
    
    def word_only(self,file):
        '''
        return only words from file
        This function is use to seperate the word from the repective file
        '''
        word = self.readFile(file)
        data = []
        for i in word:
            if i.isalpha():
                data.append(i)
        return data  
    
    def countWords(self,file):
        '''
        return the count of words int list
        '''
        words = self.word_only(file)
        word_list = {}
        for i in words:
            if i in word_list:
                word_list[i] = word_list[i] + 1
            else:
                word_list[i] = 1
        word_count = list(word_list.items())
        return word_count    
    
    def countAlphabet(self,file):
        '''
        return the count of alphabet in a file
        '''
        words_lists = self.word_only(file)
        char = {}
        for i in words_lists:
            first_char = i[0]
            if first_char in char:
                char[first_char] = char[first_char] + 1
            else:
                char[first_char] = 1
        char_count = list(char.items())       
        return char_count
    
    
    def dump_data(self,data,db_name,table_name,col1_name,col1_type,col2_name,col2_type):
        '''this function will dump the data in sqlite3 data base'''
        db = s.connect(db_name)
        c = db.cursor()
        c.execute(f'create table {table_name}({col1_name} {col1_type}, {col2_name} {col2_type})')
        for i in data:
            c.execute(f'insert into {table_name} values{i}')
        db.commit()   
        
    def printData(self,db_name,table_name):
        '''
        this function will retrive the data from data base
        '''
        db = s.connect(db_name)
        c = db.cursor()
        data = c.execute(f"select * from {table_name}")
        for i in data:
            print(i)
```
