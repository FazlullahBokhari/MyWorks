```python
import sqlite3
class BagOfWords:
    
    def readFile(self,file):
        data = open(file,'r')
        read_words = data.read()
        sep_word = read_words.split(sep='\n')
        return sep_word
    
    def countWords(self,file):
        word_list = []
        words = self.readFile(file)
        for i in set(words):
            c = 0
            for j in words:
                if i == j:
                    c += 1
                    
            word_list.append((i,c))
        return word_list  
    def countAlphabet(self,file):
        alphabet_count = []
        alphabet = self.readFile(file)
        print(alphabet)
        for i in set(alphabet):
            c = 0
            for j in alphabet:
                if i ==j:
                    c += 1
            alphabet_count.append((i,c))
        return alphabet_count    
```
